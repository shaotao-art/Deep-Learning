```python
import pandas as pd
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
x1=np.random.rand(10)
x2=np.random.rand(70)
```

```python
## input :train_acc(each batch):list
##       :valid_acc(each epoch):list
## so the length of valid_acc is less than train_acc

def plot_learning_curve(train_acc,valid_acc,error_string=None):
    x=np.arange(len(train_acc))
    x_=x[::len(train_acc)//len(valid_acc)]
    plt.title("learning curve")
    plt.plot(x,train_acc,label='train_acc')
    plt.plot(x_,valid_acc,label="valid_acc")
    plt.xlabel("batch step")
    plt.ylabel(f"error({error_string})")
    plt.legend()
    plt.show()
```

```python
plot_learning_curve(x2,x1,"MSE")
```

```python
import time
from threading import Thread
```

```python
#####    plot live learning curvel    #####

train_acc=[]
valid_acc=[]


    


def add_value():
    for i in range(0,1000):  
        time.sleep(0.5)
        train_acc.append(np.random.randn(1)[0])
        if i%3==0:
            valid_acc.append(np.random.rand(1)[0])


    



thread = Thread(target=add_value)
thread.start()

```

```python
def animate(train_acc,valid_acc,batches_in_a_epoch,error_string=None):
    x=np.arange(len(train_acc))
    x_=x[::batches_in_a_epoch]

    plt.cla()
    plt.title("learning curve")
    plt.plot(x,train_acc,label='train_acc')
    plt.plot(x_,valid_acc,label="valid_acc")
    plt.xlabel("batch step")
    plt.ylabel(f"error({error_string})")
    plt.legend()


ani=FuncAnimation(plt.gcf(),animate,interval=1000)
plt.show()
```
